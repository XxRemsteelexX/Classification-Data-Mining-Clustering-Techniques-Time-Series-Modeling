```python

#  function to check for stationarity
def check_stationarity_with_plots(series, window=30, figsize=(12, 8)):
    """
    Check stationarity of a time series using rolling statistics & statistical tests
    
    Parameters:
    -----------
    series : pandas Series
        The time series to check
    window : int, default=30
        Window size for rolling statistics
    figsize : tuple, default=(12, 8)
        Figure size for the plot
    """
    # Calc rolling statistics
    rolling_mean = series.rolling(window=window).mean()
    rolling_std = series.rolling(window=window).std()

    
    # plot rolling statistics
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=figsize)
    
    ax1.plot(series, label='Original Series')
    ax1.plot(rolling_mean, label=f'Rolling Mean (window={window})')
    ax1.plot(rolling_std, label=f'Rolling Std (window={window})')
    ax1.set_title('Rolling Mean & Standard Deviation', fontsize=16)
    ax1.grid(True, alpha=0.3)
    ax1.legend(loc='best')



    
    # plot histogram
    ax2.hist(series, bins=30, alpha=0.7)
    ax2.set_title('Histogram of Series Values', fontsize=16)
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('figures/stationarity_check.png', dpi=300)
    plt.show()
    
    # performing ADF Test
    print('\nAugmented Dickey-Fuller Test:')
    adf_result = adfuller(series.dropna())
    labels = ['ADF Test Statistic', 'p-value', 'Lags Used', 'Number of Observations Used']
    for value, label in zip(adf_result, labels):
        print(f'{label}: {value}')
    
    print('\nCritical Values:')
    for key, value in adf_result[4].items():
        print(f'   {key}: {value}')

    
    # performing KPSS Test
    print('\nKPSS Test:')
    kpss_result = kpss(series.dropna())
    labels = ['KPSS Test Statistic', 'p-value', 'Lags Used']
    for value, label in zip(kpss_result, labels):
        print(f'{label}: {value}')
    
    print('\nCritical Values:')
    for key, value in kpss_result[3].items():
        print(f'   {key}: {value}')

    
    # interpretation
    if adf_result[1] < 0.05:
        adf_conclusion = "Series is stationary (reject H0)"
    else:
        adf_conclusion = "Series is non-stationary (fail to reject H0)"
        
    if kpss_result[1] < 0.05:
        kpss_conclusion = "Series is non-stationary (reject H0)"
    else:
        kpss_conclusion = "Series is stationary (fail to reject H0)"
    
    print(f"\nADF Test Conclusion: {adf_conclusion}")
    print(f"KPSS Test Conclusion: {kpss_conclusion}")

    
```

```python

#looking at the stationarity of the original series
print("Stationarity Check for Original Series")
check_stationarity_with_plots(data['Revenue'])
```

```python

# create differenced series and check stationarity

data['Revenue_Diff'] = data['Revenue'].diff()

# checking stationarity of the differenced series
print("Stationarity Check for Differenced Series")
check_stationarity_with_plots(data['Revenue_Diff'].dropna())
```

```python

# define function for ACF and PACF analysis
def plot_acf_pacf(series, lags=40, figsize=(12, 8)):
    """
    Plot the ACF and PACF for time series
    
    Parameters:
    -----------
    series : pandas Series
        The time series to analyze
    lags : int, default=40
        Number of lags to include in the plots
    figsize : tuple, default=(12, 8)
        Figure size for the plot
    """
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=figsize)
    
    plot_acf(series.dropna(), ax=ax1, lags=lags)
    ax1.set_title('Autocorrelation Function (ACF)', fontsize=16)
    ax1.grid(True, alpha=0.3)
    
    plot_pacf(series.dropna(), ax=ax2, lags=lags)
    ax2.set_title('Partial Autocorrelation Function (PACF)', fontsize=16)
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('figures/acf_pacf.png', dpi=300)
    plt.show()
```

```python


# checking correlation
# plot ACF and PACF of the original series
print("ACF and PACF for Original Series")
plot_acf_pacf(data['Revenue'])


# plot ACF and PACF of the differenced series
print("ACF and PACF for Differenced Series")
plot_acf_pacf(data['Revenue_Diff'].dropna())
```

```python

#  function for time series decomposition
def decompose_time_series(series, period=7, model='additive', figsize=(12, 10)):
    """
    decompose a time series into trend, seasonal, and residual components
    
    Psarameters:
    -----------
    series : pandas Series
        The time series to decompose
    period : int, default=7
        The periodicity of the seasonal component
    model : str, default='additive'
        The type of decomposition ('additive' or 'multiplicative')
    figsize : tuple, default=(12, 10)
        Figure size for the plot
    """
    #  decomposition
    decomposition = seasonal_decompose(series, model=model, period=period)
    
    # plot components
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=figsize)
    
    ax1.plot(series, label='Original')
    ax1.set_title('Original Series', fontsize=14)
    ax1.legend(loc='best')
    ax1.grid(True, alpha=0.3)
    
    ax2.plot(decomposition.trend, label='Trend')
    ax2.set_title('Trend Component', fontsize=14)
    ax2.legend(loc='best')
    ax2.grid(True, alpha=0.3)
    
    ax3.plot(decomposition.seasonal, label='Seasonality')
    ax3.set_title('Seasonal Component', fontsize=14)
    ax3.legend(loc='best')
    ax3.grid(True, alpha=0.3)
    
    ax4.plot(decomposition.resid, label='Residuals')
    ax4.set_title('Residual Component', fontsize=14)
    ax4.legend(loc='best')
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('figures/decomposition.png', dpi=300)
    plt.show()
    
    # checking for stationarity in residuals
    residuals = decomposition.resid.dropna()
    print("Stationarity Check for Residuals:")
    adf_result = adfuller(residuals)
    print(f'ADF Test Statistic: {adf_result[0]}')
    print(f'p-value: {adf_result[1]}')
    
    # residual autocorrelation
    fig, ax = plt.subplots(figsize=(10, 6))
    plot_acf(residuals, ax=ax, lags=30)
    ax.set_title('ACF of Residuals', fontsize=16)
    plt.tight_layout()
    plt.savefig('figures/residual_acf.png', dpi=300)
    plt.show()
    
    return decomposition
    
```

```python

# decompose the time series ( weekly seasonality)
print("Time Series Decomposition")
decomposition = decompose_time_series(data['Revenue'], period=7)
```

```python


# split data into training and test sets (80% training, 20% testing)
train_size = int(len(data) * 0.8)
train_data = data.iloc[:train_size].copy()
test_data = data.iloc[train_size:].copy()


print(f"Training set size: {train_data.shape[0]} observations")
print(f"Test set size: {test_data.shape[0]} observations")


# droping rows with NaN values from lagged features before model training
train_data.dropna(inplace=True)
print(f"Training set size after removing NaNs: {train_data.shape[0]} observations")


# for each lag feature in test data, fill first values from the end of training data
for lag in [1, 7, 30]:
    lag_col = f'Revenue_Lag{lag}'
    # Get the last 'lag' values from training set
    last_values = data['Revenue'].iloc[train_size-lag:train_size].values
    # Fill the first 'lag' values in test set
    for j in range(min(lag, len(test_data))):
        if j < len(last_values):
            test_data[lag_col].iloc[j] = last_values[j]


# plot the split
plt.figure(figsize=(12, 6))
plt.plot(train_data['Day'], train_data['Revenue'], label='Training Data')
plt.plot(test_data['Day'], test_data['Revenue'], label='Test Data')
plt.axvline(x=train_data['Day'].iloc[-1], color='r', linestyle='--', label='Train/Test Split')
plt.title('Train-Test Split of Revenue Time Series', fontsize=16)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Revenue (Millions $)', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('figures/train_test_split.png', dpi=300)
plt.show()
```

```python
#  function for ARIMA model fitting and evaluation
def fit_evaluate_arima(p_values, d_values, q_values, train, test, seasonal=False, s_values=None):
    """
    Fit multiple ARIMA models with different parameters and evaluate them on test data
    
    Parameters:
    -----------
    p_values : list
        Autoregressive order values to try
    d_values : list
        Differencing order values to try
    q_values : list
        Moving average order values to try
    train : pandas DataFrame
        Training data
    test : pandas DataFrame
        Test data
    seasonal : bool, default=False
        Whether to include seasonal components
    s_values : list, default=None
        Seasonal orders to try if seasonal=True
    
    Returns:
    --------
    best_model : dict
        Dictionary containing information about the best model
    """
    if seasonal and s_values is None:
        s_values = [0]
        
    best_aic = float('inf')
    best_bic = float('inf')  
    best_rmse = float('inf') 
    best_model = None
    best_params = None
    results = []
    
    for p in p_values:
        for d in d_values:
            for q in q_values:
                if seasonal:
                    for s in s_values:
                        try:
                            model = SARIMAX(train['Revenue'], 
                                           order=(p, d, q),
                                           seasonal_order=(1, 0, 0, s),
                                           enforce_stationarity=False,
                                           enforce_invertibility=False)
                            fitted_model = model.fit(disp=0)
                            
                            # get AIC and BIC
                            aic = fitted_model.aic
                            bic = fitted_model.bic
                            
                            # generate forecasts for test period
                            forecast = fitted_model.get_forecast(steps=len(test))
                            forecast_values = forecast.predicted_mean
                            
                            # Calc error metrics
                            mse = mean_squared_error(test['Revenue'], forecast_values)
                            rmse = np.sqrt(mse)
                            mae = mean_absolute_error(test['Revenue'], forecast_values)
                            mape = np.mean(np.abs((test['Revenue'].values - forecast_values) / test['Revenue'].values)) * 100
                            
                            # store results
                            model_info = {
                                'order': (p, d, q),
                                'seasonal_order': (1, 0, 1, s) if seasonal else None,
                                'aic': aic,
                                'bic': bic,
                                'rmse': rmse,
                                'mae': mae,
                                'mape': mape,
                                'model': fitted_model
                            }
                            results.append(model_info)
                            
                            # update best model based on criteria:
                            # first prioritize out-of-sample RMSE
                            # if RMSE is close (within 5%), prefer simpler model using BIC
                            if rmse < best_rmse * 0.95:  # better RMSE (>5% improvement)
                                best_rmse = rmse
                                best_aic = aic
                                best_bic = bic
                                best_model = fitted_model
                                best_params = {
                                    'order': (p, d, q),
                                    'seasonal_order': (1, 0, 1, s) if seasonal else None
                                }
                            elif rmse < best_rmse * 1.05:  # RMSE within 5% of best
                                if bic < best_bic:  # select simpler model using BIC
                                    best_rmse = rmse
                                    best_aic = aic
                                    best_bic = bic
                                    best_model = fitted_model
                                    best_params = {
                                        'order': (p, d, q),
                                        'seasonal_order': (1, 0, 1, s) if seasonal else None
                                    }
                            
                            print(f"ARIMA{(p,d,q)}" + (f"x(1,0,1,{s})" if seasonal else "") + 
                                  f" - AIC: {aic:.2f}, BIC: {bic:.2f}, RMSE: {rmse:.2f}, MAPE: {mape:.2f}%")
                            
                        except Exception as e:
                            print(f"Error fitting ARIMA{(p,d,q)}" + (f"x(1,0,1,{s})" if seasonal else "") + f": {str(e)}")
                else:
                    try:
                        model = ARIMA(train['Revenue'], order=(p, d, q))
                        fitted_model = model.fit()
                        
                        #AIC and BIC
                        aic = fitted_model.aic
                        bic = fitted_model.bic
                        
                        # generate forecasts for test period
                        forecast = fitted_model.get_forecast(steps=len(test))
                        forecast_values = forecast.predicted_mean
                        
                        # Calc error metrics
                        mse = mean_squared_error(test['Revenue'], forecast_values)
                        rmse = np.sqrt(mse)
                        mae = mean_absolute_error(test['Revenue'], forecast_values)
                        mape = np.mean(np.abs((test['Revenue'].values - forecast_values) / test['Revenue'].values)) * 100
                        
                        # store results
                        model_info = {
                            'order': (p, d, q),
                            'seasonal_order': None,
                            'aic': aic,
                            'bic': bic,
                            'rmse': rmse,
                            'mae': mae,
                            'mape': mape,
                            'model': fitted_model
                        }
                        results.append(model_info)
                        
                        # update best model based on criteria
                        if rmse < best_rmse * 0.95:  #  better RMSE (>5% improvement)
                            best_rmse = rmse
                            best_aic = aic
                            best_bic = bic
                            best_model = fitted_model
                            best_params = {
                                'order': (p, d, q),
                                'seasonal_order': None
                            }
                        elif rmse < best_rmse * 1.05:  # RMSE within 5% of best
                            if bic < best_bic:  # select simpler model using BIC
                                best_rmse = rmse
                                best_aic = aic
                                best_bic = bic
                                best_model = fitted_model
                                best_params = {
                                    'order': (p, d, q),
                                    'seasonal_order': None
                                }
                        
                        print(f"ARIMA{(p,d,q)} - AIC: {aic:.2f}, BIC: {bic:.2f}, RMSE: {rmse:.2f}, MAPE: {mape:.2f}%")
                        
                    except Exception as e:
                        print(f"Error fitting ARIMA{(p,d,q)}: {str(e)}")
    
    # sort results by RMSE primary and BIC secondary
    results.sort(key=lambda x: (x['rmse'], x['bic']))
    
    return {
        'best_model': best_model,
        'best_params': best_params,
        'best_aic': best_aic,
        'best_bic': best_bic,
        'best_rmse': best_rmse,
        'all_results': results
    }
```

```python
#fit ARIMA models with different parameters

# define parameter ranges to try
p_values = [0, 1, 2]
d_values = [1]  #  determined that d=1 is appropriate
q_values = [0, 1, 2]
s_values = [7, 14, 30]  # test weekly bi-weekly and monthly seasonality

# find the best ARIMA model
print("Fitting ARIMA models...")
arima_results = fit_evaluate_arima(p_values, d_values, q_values, train_data, test_data)
```

```python
#fit Seasonal ARIMA models
# find the best SARIMA model 

print("Fitting SARIMA models...")
sarima_results = fit_evaluate_arima(p_values, d_values, q_values, train_data, test_data, seasonal=True, s_values=s_values)
```

```python

# determine the overall best model
print("\nModel Comparison:")
print(f"Best ARIMA: {arima_results['best_params']['order']} - RMSE: {arima_results['best_rmse']:.4f}, BIC: {arima_results['best_bic']:.2f}")
if sarima_results['best_params']['seasonal_order']:
    print(f"Best SARIMA: {sarima_results['best_params']['order']}x{sarima_results['best_params']['seasonal_order']} - RMSE: {sarima_results['best_rmse']:.4f}, BIC: {sarima_results['best_bic']:.2f}")

# Calc improvement percentage
rmse_improvement = (arima_results['best_rmse'] - sarima_results['best_rmse']) / arima_results['best_rmse'] * 100

# determine the overall best model based on both parsimony and performance
if sarima_results['best_rmse'] < arima_results['best_rmse'] * 0.95: 
    best_results = sarima_results
    model_type = 'SARIMA'
    print(f"\nSelected SARIMA model (RMSE {rmse_improvement:.2f}% better than ARIMA)")
elif sarima_results['best_rmse'] < arima_results['best_rmse']:  
    if sarima_results['best_bic'] < arima_results['best_bic']:  
        best_results = sarima_results
        model_type = 'SARIMA'
        print(f"\nSelected SARIMA model (slightly better RMSE and better BIC)")
    else:
        best_results = arima_results
        model_type = 'ARIMA'
        print(f"\nSelected ARIMA model (SARIMA only marginally better but more complex)")
else:
    best_results = arima_results
    model_type = 'ARIMA'
    print(f"\nSelected ARIMA model (better RMSE)")

best_model = best_results['best_model']
best_params = best_results['best_params']

print(f"\nBest model: {model_type} with parameters {best_params}")
print(f"AIC: {best_results['best_aic']:.2f}")
print(f"BIC: {best_results['best_bic']:.2f}")
print(f"RMSE: {best_results['best_rmse']:.4f}")

# print model summary
print("\nModel Summary:")
print(best_model.summary())
```

```python


# Calc error metrics
mse = mean_squared_error(forecast_df['actual'], forecast_df['forecast'])
rmse = np.sqrt(mse)
mae = mean_absolute_error(forecast_df['actual'], forecast_df['forecast'])
mape = np.mean(np.abs((forecast_df['actual'] - forecast_df['forecast']) / forecast_df['actual'])) * 100

print(f"\nForecast Error Metrics:")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")

# Calc mean directional accuracy (MDA)
actual_diff = np.diff(forecast_df['actual'])
forecast_diff = np.diff(forecast_df['forecast'])
correct_direction = (actual_diff * forecast_diff) > 0
mda = np.mean(correct_direction) * 100

print(f"Mean Directional Accuracy (MDA): {mda:.2f}%")
```

```python
# analyze MDA in more detail
windows = [1, 3, 7, 14]  # analyze directional accuracy over different windows
mda_results = {}

for window in windows:
    # Calc rolling direction for actuals and forecast
    actual_direction = np.sign(np.diff(forecast_df['actual'].rolling(window=window).mean()))
    forecast_direction = np.sign(np.diff(forecast_df['forecast'].rolling(window=window).mean()))
    
    # filter out zero changes no direction
    mask = (actual_direction != 0)
    if sum(mask) > 0:
        correct = (actual_direction[mask] == forecast_direction[mask])
        window_mda = np.mean(correct) * 100
        mda_results[window] = window_mda
    else:
        mda_results[window] = np.nan

print("\nMean Directional Accuracy (MDA) over different windows:")
for window, accuracy in mda_results.items():
    print(f"{window}-day window: {accuracy:.2f}%")

# Viz directional accuracy
plt.figure(figsize=(12, 6))
plt.bar(list(mda_results.keys()), list(mda_results.values()), alpha=0.7)
plt.axhline(y=50, color='r', linestyle='--', label='Random Guess (50%)')
plt.title('Mean Directional Accuracy Over Different Time Windows', fontsize=16)
plt.xlabel('Window Size (Days)', fontsize=12)
plt.ylabel('Directional Accuracy (%)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()
plt.savefig('figures/directional_accuracy.png', dpi=300)
plt.show()

```

```python

# time series cross-validation function
def time_series_cv(data, model_type, model_params, n_splits=5):
    """
    Perform time series cross-validation to assess model generalization
    
    Parameters:
    -----------
    data : pandas DataFrame
        The full dataset
    model_type : str
        'ARIMA' or 'SARIMA'
    model_params : dict
        Model parameters including 'order' and 'seasonal_order'
    n_splits : int, default=5
        Number of cross-validation splits
    
    Returns:
    --------
    cv_results : dict
        Cross-validation metrics
    """
    # Init results storage
    cv_rmse = []
    cv_mape = []
    
    # determine initial training size at least 30% of data
    initial_train_size = max(int(len(data) * 0.3), 90)  # at least 90 days or 30%
    
    # Calc the size of each testing fold
    remaining_data = len(data) - initial_train_size
    fold_size = remaining_data // n_splits
    
    for i in range(n_splits):
        # set up training and validation sets for this fold
        train_end = initial_train_size + i * fold_size
        val_start = train_end
        val_end = val_start + fold_size
        
        # extract training and validation sets
        train_fold = data.iloc[:train_end].copy()
        val_fold = data.iloc[val_start:val_end].copy()
        
        try:
            # fit model based on type
            if model_type == 'ARIMA':
                model = ARIMA(train_fold['Revenue'], order=model_params['order'])
                fitted_model = model.fit()
            else:  # SARIMA
                model = SARIMAX(train_fold['Revenue'], 
                               order=model_params['order'],
                               seasonal_order=model_params['seasonal_order'],
                               enforce_stationarity=False,
                               enforce_invertibility=False)
                fitted_model = model.fit(disp=0)
            
            # forecast validation period
            forecast = fitted_model.get_forecast(steps=len(val_fold))
            forecast_values = forecast.predicted_mean
            
            # Calc metrics
            rmse = np.sqrt(mean_squared_error(val_fold['Revenue'], forecast_values))
            mape = np.mean(np.abs((val_fold['Revenue'].values - forecast_values) / val_fold['Revenue'].values)) * 100
            
            cv_rmse.append(rmse)
            cv_mape.append(mape)
            
            print(f"Fold {i+1}: RMSE = {rmse:.4f}, MAPE = {mape:.2f}%")
            
        except Exception as e:
            print(f"Error in fold {i+1}: {str(e)}")
    
    # Calc average metrics
    avg_rmse = np.mean(cv_rmse)
    avg_mape = np.mean(cv_mape)
    std_rmse = np.std(cv_rmse)
    
    return {
        'fold_rmse': cv_rmse,
        'fold_mape': cv_mape,
        'avg_rmse': avg_rmse,
        'avg_mape': avg_mape,
        'std_rmse': std_rmse
    }

# perform cross-validation for both models
print("Cross-validation for ARIMA model:")
arima_cv = time_series_cv(data, 'ARIMA', arima_results['best_params'], n_splits=3)

print("\nCross-validation for SARIMA model:")
sarima_cv = time_series_cv(data, 'SARIMA', sarima_results['best_params'], n_splits=3)

# compare cross-validation results
print("\nCross-validation comparison:")
print(f"ARIMA: Avg RMSE = {arima_cv['avg_rmse']:.4f} (±{arima_cv['std_rmse']:.4f}), Avg MAPE = {arima_cv['avg_mape']:.2f}%")
print(f"SARIMA: Avg RMSE = {sarima_cv['avg_rmse']:.4f} (±{sarima_cv['std_rmse']:.4f}), Avg MAPE = {sarima_cv['avg_mape']:.2f}%")

# overfitting check, if test RMSE is much worse than training RMSE
print("\nOverfitting check:")
print(f"ARIMA - CV RMSE vs Test RMSE: {arima_cv['avg_rmse']:.4f} vs {arima_results['best_rmse']:.4f}")
print(f"SARIMA - CV RMSE vs Test RMSE: {sarima_cv['avg_rmse']:.4f} vs {sarima_results['best_rmse']:.4f}")

```
