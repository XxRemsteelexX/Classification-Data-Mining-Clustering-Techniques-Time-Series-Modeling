```python

#  function to check for stationarity
def check_stationarity_with_plots(series, window=30, figsize=(12, 8)):
    """
    Check stationarity of a time series using rolling statistics & statistical tests
    
    Parameters:
    -----------
    series : pandas Series
        The time series to check
    window : int, default=30
        Window size for rolling statistics
    figsize : tuple, default=(12, 8)
        Figure size for the plot
    """
    # Calc rolling statistics
    rolling_mean = series.rolling(window=window).mean()
    rolling_std = series.rolling(window=window).std()

    
    # plot rolling statistics
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=figsize)
    
    ax1.plot(series, label='Original Series')
    ax1.plot(rolling_mean, label=f'Rolling Mean (window={window})')
    ax1.plot(rolling_std, label=f'Rolling Std (window={window})')
    ax1.set_title('Rolling Mean & Standard Deviation', fontsize=16)
    ax1.grid(True, alpha=0.3)
    ax1.legend(loc='best')



    
    # plot histogram
    ax2.hist(series, bins=30, alpha=0.7)
    ax2.set_title('Histogram of Series Values', fontsize=16)
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('figures/stationarity_check.png', dpi=300)
    plt.show()
    
    # performing ADF Test
    print('\nAugmented Dickey-Fuller Test:')
    adf_result = adfuller(series.dropna())
    labels = ['ADF Test Statistic', 'p-value', 'Lags Used', 'Number of Observations Used']
    for value, label in zip(adf_result, labels):
        print(f'{label}: {value}')
    
    print('\nCritical Values:')
    for key, value in adf_result[4].items():
        print(f'   {key}: {value}')

    
    # performing KPSS Test
    print('\nKPSS Test:')
    kpss_result = kpss(series.dropna())
    labels = ['KPSS Test Statistic', 'p-value', 'Lags Used']
    for value, label in zip(kpss_result, labels):
        print(f'{label}: {value}')
    
    print('\nCritical Values:')
    for key, value in kpss_result[3].items():
        print(f'   {key}: {value}')

    
    # interpretation
    if adf_result[1] < 0.05:
        adf_conclusion = "Series is stationary (reject H0)"
    else:
        adf_conclusion = "Series is non-stationary (fail to reject H0)"
        
    if kpss_result[1] < 0.05:
        kpss_conclusion = "Series is non-stationary (reject H0)"
    else:
        kpss_conclusion = "Series is stationary (fail to reject H0)"
    
    print(f"\nADF Test Conclusion: {adf_conclusion}")
    print(f"KPSS Test Conclusion: {kpss_conclusion}")

    
```

```python

#looking at the stationarity of the original series
print("Stationarity Check for Original Series")
check_stationarity_with_plots(data['Revenue'])
```

```python

# create differenced series and check stationarity

data['Revenue_Diff'] = data['Revenue'].diff()

# checking stationarity of the differenced series
print("Stationarity Check for Differenced Series")
check_stationarity_with_plots(data['Revenue_Diff'].dropna())
```
